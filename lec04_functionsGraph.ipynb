#### Loop through functions inputs


```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'r') # red x axis
plt.plot([0,0],[ymin,ymax], 'r') # red y axis
print('x \t y')
for x in range(10):
    y = 0.5*x + 1
    plt.plot([x],[y], 'bo')
    print(x,'\t',y)

plt.show()
```

### Use Array as an input

The line x = np.linspace(xmin, xmax, points) uses the numpy library to generate an array of evenly spaced numbers over a specified range. Here's a breakdown of what it does:

#### Components
np.linspace(start, stop, num): This function returns num evenly spaced samples, calculated over the interval [start, stop].
##### xmin: The starting value of the sequence.
##### xmax: The end value of the sequence.
##### points: The number of samples to generate between xmin and xmax.
#### Example Explanation
If xmin = -10, xmax = 10, and points = 100, the code will generate an array of 100 evenly spaced numbers from -10 to 10. The resulting array x will include values such as -10, -9.7979798, -9.5959596, ..., 9.7979798, 10.

This line is often used in plotting functions to create a smooth curve by providing a set of x-coordinates at which the function will be evaluated. For example, if you want to plot a function 

(

)
f(x), you can use the values in x to calculate the corresponding y values and plot the curve.

```python
import numpy as np
xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2*(xmax-xmin) 
x = np.linspace(xmin, xmax, points)
print(x,len(x))

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'r') # red x axis
plt.plot([0,0],[ymin,ymax], 'r') # red y axis

y = 2*x +1
plt.plot(x,y, 'bo')

plt.show()
```

### Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10
points = 2*(xmax-xmin)
x = np.linspace(xmin, xmax, points)

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

ax.set_xlabel("x values") #set the x label
ax.set_ylabel("y values") #set the y label
ax.set_title("Graph Title") #set the title
ax.grid(True) #set the grid

ax.set_xticks(np.arange(xmin, xmax+1, 1)) 
ax.set_yticks(np.arange(ymin, ymax+1, 1))

y = 2*x +1
plt.plot(x,y, label='y=2x+1')
plt.plot([4],[6], 'ro', label='point')
plt.plot(x,3*x, label='steeper line')
plt.legend()
plt.show()
```

The lines ax.set_xticks(np.arange(xmin, xmax, 1)) and ax.set_yticks(np.arange(ymin, ymax, 1)) set the ticks on the x and y axes of the plot. Here's what each part does:

### Components
#### ax: This is the Axes object created by plt.subplots(). It represents the part of the figure that contains the data space, where the data is plotted.

#### set_xticks(): This method sets the positions of the ticks on the x-axis.

#### set_yticks(): This method sets the positions of the ticks on the y-axis.

#### np.arange(start, stop, step): This function returns an array of values starting from start up to (but not including) stop, with a step size of step.

### Specific Usage
#### ax.set_xticks(np.arange(xmin, xmax, 1)):

#### np.arange(xmin, xmax, 1) generates an array of values starting from xmin to xmax - 1 with a step of 1.
#### `These values are used as the positions for the ticks on the x-axis.
For example, if xmin = -10 and xmax = 10, np.arange(-10, 10, 1) will produce [-10, -9, -8, ..., 8, 9]. These values will be the positions of the ticks on the x-axis.

#### ax.set_yticks(np.arange(ymin, ymax, 1)):

np.arange(ymin, ymax, 1) generates an array of values starting from ymin to ymax - 1 with a step of 1.
These values are used as the positions for the ticks on the y-axis.
For example, if ymin = -10 and ymax = 10, np.arange(-10, 10, 1) will produce [-10, -9, -8, ..., 8, 9]. These values will be the positions of the ticks on the y-axis.
