#### Loop through functions inputs


```python
import matplotlib.pyplot as plt

xmin = -10
xmax = 10
ymin = -10
ymax = 10

fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'r') # red x axis
plt.plot([0,0],[ymin,ymax], 'r') # red y axis
print('x \t y')
for x in range(10):
    y = 0.5*x + 1
    plt.plot([x],[y], 'bo')
    print(x,'\t',y)

plt.show()
```

#### This code uses the matplotlib library to create a simple plot with Cartesian coordinates and plots a series of points along a line. Here's a breakdown of what each part does:
### Importing the library

```python
import matplotlib.pyplot as plt
```

#### This imports the pyplot module from the matplotlib library, which provides functions for plotting data.

### Setting the Plot Limits:

```python
xmin = -10
xmax = 10
ymin = -10
ymax = 10
```

#### These variables define the range for the x and y axes.
### Creating the Figure and Axes:
#### This line creates a figure and a set of axes. fig is the Figure object, and ax is the Axes object.


```python
fig, ax = plt.subplots()
```

### Setting the Axis Limits:
#### This sets the limits for the x-axis and y-axis according to the values of xmin, xmax, ymin, and ymax. The plot will display this range.

```python
plt.axis([xmin, xmax, ymin, ymax])
```

### Drawing the X and Y Axes:
#### These two lines plot the x-axis and y-axis in red ('r'). The first line draws the x-axis from xmin to xmax at y=0. The second line draws the y-axis from ymin to ymax at x=0.

```python
plt.plot([xmin, xmax], [0, 0], 'r')  # red x-axis
plt.plot([0, 0], [ymin, ymax], 'r')  # red y-axis
```

### Plotting Points:
#### This loop iterates over the range of x values from 0 to 9. For each x value, it calculates the corresponding y value using the equation y = 0.5 * x + 1. The plt.plot([x], [y], 'bo') line plots each point (x, y) with blue circles ('bo').

```python
plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0, 0], 'r')  # red x-axis
plt.plot([0, 0], [ymin, ymax], 'r')  # red y-axis
for x in range(10):
    y = 0.5 * x + 1
    plt.plot([x], [y], 'bo')
    
plt.show()
```

### Displaying the Plot:
#### This line displays the plot window.

```python
plt.show()
```
