#### A town population increases by constant rate. In 2010 the population was 55000. By 2012 the population had increased to 76000. If this trend continues the predict the population in 2016.

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = 0
y1 = 55
x2 = 2
y2 = 76

# Develop the equation y = mx + b
m = (y2 - y1) / (x2 - x1)
b = y1 - m*x1
print("y = ", m, "x + ", b)


# For the graph
xmin = 0
xmax = 10
ymin = 0
ymax = 150
points = 2*(xmax-xmin)
x = np.linspace(xmin, xmax, points)

# For the line on the graph


y = m*x + b
  
# Basic setup for the graph
fig, ax = plt.subplots()
plt.axis([xmin,xmax,ymin,ymax]) # window size
plt.plot([xmin,xmax],[0,0],'b') # blue x axis
plt.plot([0,0],[ymin,ymax], 'b') # blue y axis

# Add details to the graph
ax.set_xlabel("Time (years)")
ax.set_ylabel("Population (K)")
ax.grid(True)
ax.set_title("Rate of Population") #set the title
#ax.set_xticks(np.arange(xmin, xmax, 2))
#ax.set_yticks(np.arange(ymin, ymax, 1))


# Plot the linear function as a red line
plt.plot(x,y,'r')


plt.show()

y= m*8 +b
print("Population in 2016 is ",y,"K")
```
